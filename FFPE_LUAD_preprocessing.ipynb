# Load SS2 human LUAD data (feather objects)

```python
# SS2
ad_ss2_old = ffpe.import_feather('20191030_lung_brain_hg19_mm10',"ss2","191030")
ad_ss2_old.obs['library_type'] = 'FFPE_SS2'
LUAD_names = [n for n in ad_ss2_old.obs_names if 'huLUAD' in n]

ad_ss2_old = ad_ss2_old[LUAD_names,:].copy()
ad_ss2_old.obs['plate'] = '191030'

ad_ss2_new = ffpe.import_feather('20210413_AM_lung_SS2',"ss2","210413")
ad_ss2_new.obs['library_type'] = 'FFPE_SS2'
ad_ss2 = ad_ss2_old.concatenate(ad_ss2_new, join='outer')
```

```python
ad_ss2.X[np.isnan(ad_ss2.X)] = 0
ad_ss2.obs['nuclei_ct'] = ad_ss2.obs['num_nuclei'].astype('category')
```

### split by hg19 vs mm10

```python
ad_single = ad_ss2[ad_ss2.obs['num_nuclei']==1].copy()
ad_hg19, ad_mm10, ad_all = ffpe.split_hg19_and_mm10(ad_single)
```

```python
ad_mm10 = ffpe.calc_mito_ncounts(ad_mm10, case='mouse')
ad_hg19 = ffpe.calc_mito_ncounts(ad_hg19, case='human')
ad_all = ffpe.calc_mito_ncounts(ad_all, case='hg19')
```

```python
ad_all.obs['log10_ncounts'] = np.log10(ad_all.obs['n_counts_hg19'])
ad_all.obs['log10_ngenes'] = np.log10(ad_all.obs['n_genes_hg19'])
```

```python
ffpe.scatter_hg19_vs_mm10(ad_all, hue_var='plate', savefig=True)
```

```python
sc.pl.violin(ad_all, keys=['percent_hg19'], groupby='plate', palette=['#bcbcbc'], save='_perc_hg19')
sc.pl.violin(ad_all, keys=['percent_mm10'], groupby='plate', palette=['#bcbcbc'], save='_perc_mm10')
```

```python
# keep nuclei with mitochondrial content less than 10%
print(ad_hg19.shape)
ad_hg19 = ad_hg19[ad_hg19.obs['frac_mito']<0.1,:].copy()
print(ad_hg19.shape)
```

```python
from scipy.sparse import csr_matrix, csc_matrix
a = csr_matrix(ad_hg19.X)
ad_hg19.X = a
ad_hg19.layers['counts'] = ad_hg19.X.copy()
```

```python
ad_hg19.obs.drop(columns=['i7_bc','i5_bc','i7_rc','i5_rc'], inplace=True)
```

```python
ad_hg19.obs['log10_ncounts'] = ad_all.obs['log10_ncounts']
ad_hg19.obs['log10_ngenes'] = ad_all.obs['log10_ngenes']
```

```python
ad_hg19.obs['threshold'] = False
ad_hg19.obs.at[((ad_hg19.obs['n_genes_hg19']>MIN_GENES)&(ad_hg19.obs['n_genes_hg19']<MAX_GENES)),'threshold'] = True

print(ad_hg19.obs.shape)
print(sum(ad_hg19.obs['threshold']))

ad_hg19_thresh = ad_hg19[ad_hg19.obs['threshold'],:].copy()
```

```python
def subset_genes(ad):
    print(ad.shape)
    ad = ad[:,ad.layers['counts'].sum(axis=0).flatten()>0].copy()
    print(ad.shape)
    return ad

ad_hg19 = subset_genes(ad_hg19)
```

```python
sns.scatterplot(data=ad_hg19_thresh.obs, 
                x='log10_ncounts', y='log10_ngenes', 
                hue='plate', size=50)
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.savefig('%s/scatter_log10_ncounts_ngenes_ALL.pdf' %(sc.settings.figdir), bbox_inches='tight')
```

```python
sc.pl.violin(ad_hg19_thresh, keys=['n_genes_hg19'], groupby='plate')
```

```python
ad_hg19 = ad_hg19_thresh.copy()
ad_hg19.X
```

Remove bad genes

```python
# remove RP11 genes
print(ad_hg19.shape)
RP11 = ad_hg19.var_names.str.startswith('RP11-')
print('%i RP11 genes' %sum(RP11))
ad_hg19 = ad_hg19[:,~RP11].copy()
LINC = ad_hg19.var_names.str.startswith('LINC')
print('%i LINC genes' %sum(LINC))
ad_hg19 = ad_hg19[:,~LINC].copy()
ad_hg19.obs['n_genes'] = ad_hg19.X.toarray().astype(bool).sum(axis=1)
print(ad_hg19.shape)
```

```python
ad_hg19 = subset_genes(ad_hg19)
print(ad_hg19.shape)
```

```python
ad_hg19.layers['log_cts'] = ad_hg19.layers['counts'].copy()
sc.pp.normalize_total(ad_hg19, layer='log_cts')
sc.pp.log1p(ad_hg19, layer='log_cts')

sc.pp.normalize_total(ad_hg19)
sc.pp.log1p(ad_hg19)
ad_hg19.layers['scaled'] = ad_hg19.X.copy()
```
