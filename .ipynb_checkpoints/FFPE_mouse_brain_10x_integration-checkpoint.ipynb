# Load all 10x data (FFPE and fresh) and combine with SS2

```python
CELLBENDER = False
sample_dirs = ['20200216_FFPE_10x_hiseq',
           'habib_cortex', 
           'habib_cortex', 
           'habib_cortex', 
           'habib_cortex']
samples = ['FFPE_brain_protK',
           'CRTX_4mWT', 
           'CRTX_6wWT', 
           'CRTX_7mWT',
           'CRTX_10mWT']

# combine across batches
adatas = []

for sampdir, samp in zip(sample_dirs, samples):
    print(samp)
    
    # demultiplexed gene expression
    if CELLBENDER and (samp.endswith('left') or samp.endswith('right')):
        filename = 'FFPE/seq_data/%s/%s/cellbender_low_thresh/%s_out.h5' %(sampdir, samp, samp)
    else: 
        filename = 'FFPE/seq_data/%s/%s/raw_feature_bc_matrix.h5' %(sampdir, samp)
    ad = sc.read_10x_h5(filename)
    ad.var_names_make_unique()
    
    if 'FFPE' in samp: 
        print('FFPE in sample name')
        ad.obs['library_type'] = 'FFPE_10x'
    else:
        ad.obs['library_type'] = 'snRNA'
    
    adatas.append(ad)
adata = adatas[0].concatenate(adatas[1:],join='outer')
adata.var_names_make_unique()
```

```python
sc.pp.filter_cells(adata, min_genes=1)
sc.pp.filter_genes(adata, min_cells=2)
adata.obs
```

```python
adata.obs['n_counts'] = adata.X.sum(axis=1).flatten().tolist()[0]

# calculate mitochondrial fraction
mito_genes = adata.var_names.str.startswith('mt-')
adata.obs['frac_mito'] = (adata[:, mito_genes].X.sum(axis=1).flatten().tolist()[0])/adata.obs['n_counts']
```

```python
# add n_genes
adata.obs['n_genes'] = adata.X.astype(bool).sum(axis=1).flatten().tolist()[0]
```

```python
def get_cluster_proportions(adata,
                            cluster_key="leiden",
                            sample_key="library_type",
                            drop_values=None):
    """
    Input
    =====
    adata : AnnData object
    cluster_key : key of `adata.obs` storing cluster info
    sample_key : key of `adata.obs` storing sample/replicate info
    drop_values : list/iterable of possible values of `sample_key` that you don't want
    
    Returns
    =======
    pd.DataFrame with samples as the index and clusters as the columns and 0-100 floats
    as values
    """
    
    adata_tmp = adata.copy()
    sizes = adata_tmp.obs.groupby([cluster_key, sample_key]).size()
    props = sizes.groupby(level=1).apply(lambda x: 100 * x / x.sum()).reset_index() 
    props = props.pivot(columns=sample_key, index=cluster_key).T
    props.index = props.index.droplevel(0)
    props.fillna(0, inplace=True)
    
    if drop_values is not None:
        for drop_value in drop_values:
            props.drop(drop_value, axis=0, inplace=True)
    return props


def plot_cluster_proportions(cluster_props, 
                             cluster_palette=None,
                             xlabel_rotation=0,
                            stacked=True): 
    import seaborn as sns
    fig, ax = plt.subplots(dpi=300)
    fig.patch.set_facecolor("white")
    
    cmap = None
    if cluster_palette is not None:
        cmap = sns.palettes.blend_palette(
            cluster_palette, 
            n_colors=len(cluster_palette), 
            as_cmap=True)
   
    cluster_props.plot(
        kind="bar", 
        stacked=stacked, 
        ax=ax, 
        legend=None, 
        colormap=cmap
    )
    
    ax.legend(bbox_to_anchor=(1.01, 1), frameon=False, title="Library type")
    sns.despine(fig, ax)
    ax.tick_params(axis="x", rotation=xlabel_rotation)
    ax.set_xlabel(cluster_props.index.name.capitalize())
    ax.set_ylabel("% of nuclei in cluster")
    ax.set_xticklabels(cluster_props.index, rotation = 90)
    ax.set_yticks([0,25,50])
    ax.set_yticklabels([0,25,50])
    
    fig.tight_layout()
    
    return fig
```

```python
ad_props = get_cluster_proportions(adata, cluster_key='annotation')
fig = plot_cluster_proportions(ad_props.T, stacked=False)
fig.savefig('%s/barplot_cluster_proportion_of_library_type_grouped.pdf' %sc.settings.figdir, bbox_inches='tight')
```
