```python
# c4.cgn: cancer gene neighborhoods
# c6: oncogenic

pathway_genes = ffpe.get_pathway_genes(pathway_type=['c4.cgn','c6'])
ad_path, pathway_dict, pathways = ffpe.get_pathways_anndata(ad_ss2, pathway_genes, 
                                                                   MIN_FRAC=0.5, 
                                                                   MIN_GENES=20, 
                                                                  MAX_GENES=150,
                                                                  weighted=False)
print(len(pathways))
print(ad_path.shape)
ad_path_orig = ad_path.copy()
```

```python
fh = open('%s/%s.csv' %(sc.settings.figdir, 'Expressed_c4cgn_c6_pathways'),'w')
for k, v in pathway_genes.items(): 
    L = '%s,%s\n' %(k, ','.join([g for g in v]) )
    fh.writelines(L)
fh.close()
```

### Make pathway membership df

```python
flatten_genes = list(set([g for gl in pathway_dict.values() for g in gl]))

pathway_membership = pd.DataFrame(0, index=flatten_genes, columns=pathways)
for path in pathways: 
    pathway_membership.loc[pathway_dict[path]] = 1
```

### Calculate Jaccard similarity

```python
J_path_ij, J_values = ffpe.calculate_jaccard_pathways(pathways, pathway_dict)
```

```python
cg = ffpe.cluster_path_Jij(J_path_ij)
path_idx, Z = ffpe.extract_cluster_feats(cg)

reordered_paths = [pathways[i] for i in path_idx]
```

```python
cluster_assign = ffpe.cut_tree(Z, pathways, CUT_THRESHOLD=2.2)
```

```python
JACCARD_THRESH = 0.15

sns.reset_orig()
assignments = cluster_assign['cluster'].values
cluster_assign['cluster_reduced'] = 0
for clust in set(assignments): 
    clust_paths = cluster_assign[cluster_assign['cluster']==clust].index
    clust_path_Js = J_path_ij.loc[clust_paths][clust_paths]
    vals = clust_path_Js.values.flatten()
    valid_vals = vals[vals>0]
    median_val = np.median(valid_vals)
    
    plt.figure(figsize=(3,3))
    plt.hist(valid_vals, bins=30)
    plt.xlim([0,1])
    plt.axvline(median_val, color='k', linestyle='--')
    plt.axvline(JACCARD_THRESH, color='r')
    plt.title('Cluster %i' %clust)
    plt.xlabel('Jaccard distance')
    plt.ylabel('# Pathway pairs')
    plt.savefig('%s/pathway_dendrogram_cluster%i_pairwise_Jaccard.pdf' %(sc.settings.figdir, clust), 
               bbox_inches='tight')
    
    if not np.isnan(median_val) and median_val>=JACCARD_THRESH:
        print('Cluster %i' %clust)
        cluster_assign.at[clust_paths,'cluster_reduced'] = clust
        print(median_val)
```

```python
path_colors, c_dict = ffpe.make_pathway_cmap(cluster_assign, pathways)
```

```python
plt.figure(figsize=(20,20))
sns.set(font_scale=0.2)
cg = sns.clustermap(J_path_ij,
                    yticklabels=False,
                    xticklabels=False,
                    row_colors=path_colors,
                    vmin=0,
                    vmax=1,
                    method='ward',
                    cmap='Blues')
cg.ax_col_dendrogram.set_visible(False)
plt.savefig('%s/pathway_dendrogram_redundant_paths_colored.pdf' %(sc.settings.figdir),
                   bbox_inches='tight')
```

```python
pathways_keep = list(cluster_assign[cluster_assign['cluster_reduced']==0].index)
print(len(pathways_keep))

clust_centroid = {}
for clust in set(cluster_assign['cluster_reduced']): 
    if clust is not 0: 
        group_paths = list(cluster_assign[cluster_assign['cluster_reduced']==clust].index)
        group_centroid = J_path_ij.loc[group_paths][group_paths].median().idxmax()
        clust_centroid[clust] = group_centroid
        print(clust, group_centroid)
        pathways_keep.append(group_centroid)
        
print(len(pathways_keep))
```

```python
cluster_assign['pathway_rep'] = cluster_assign.iloc[path_idx]['cluster_reduced'].map(clust_centroid)
cluster_assign['pathway_rep'] = cluster_assign['pathway_rep'].fillna(cluster_assign.index.to_series())
```

```python
cluster_assign.iloc[path_idx].to_csv('%s/pathways_dendrogram_order_reduced_pathway.csv' %sc.settings.figdir)
```

## GO pathways

```python
def get_pathway_genes(pathway_marker_dict, pathway_dict, cluster): 
    gene_set = []
    for pathway in pathway_marker_dict[cluster]: 
        gene_set.extend(pathway_dict[pathway])
    unique_genes = list(set(gene_set))
    print(len(unique_genes))
    return unique_genes

def parse_GO_query(gene_list, species, db_to_keep='all'): 
    if db_to_keep=='all': 
#         db_to_keep = ['GO:MF', 'KEGG', 'TF', 'GO:BP']
        db_to_keep = ['TF', 'KEGG']
    GO_df = sc.queries.enrich(gene_list, org=species)
    GO_df = GO_df[GO_df['significant']==True]
    GO_df = GO_df[GO_df['source'].isin(db_to_keep)]
    return GO_df
```
